# neural network model

```python
class TextSimpleClassifier(nn.Module):

    def __init__(self, num_inputs, num_hidden, num_outputs):
        super().__init__()
        self.relu_fn = nn.ReLU()
        self.act_fn = nn.Tanh()
        self.linear1 = nn.Linear(num_inputs, num_hidden)
        self.linear2 = nn.Linear(num_hidden, num_outputs)

    def forward(self, x):
        
        x = self.linear1(x)
        x = self.act_fn(x)
        x = self.linear2(x)
        x = self.relu_fn(x)
        return x
    
text_model = TextSimpleClassifier(9,5,1)



class ImageClassifier(nn.Module):

    def __init__(self, num_inputs, num_hidden, num_outputs):
        super().__init__()
        self.relu_fn = nn.ReLU()
        self.act_fn = nn.Tanh()
        self.linear1 = nn.Linear(num_inputs, num_hidden)
        
        self.linear2 = nn.Linear(num_hidden, num_outputs)

    def forward(self, x):
        # Perform the calculation of the model to determine the prediction
        x = self.linear1(x)
        x = self.act_fn(x)
        x = self.linear2(x)
        x = self.relu_fn(x)
        return x

image_model = ImageClassifier(2048,5,1)


class CombinedClassifier(nn.Module):

    def __init__(self, size_text, size_image, num_hidden, num_outputs):
        super().__init__()
        self.text_classifier = TextSimpleClassifier(size_text, num_hidden, num_outputs)
        self.image_classifier = ImageClassifier(size_image, num_hidden, num_outputs)
        self.linear1 = nn.Linear(num_outputs*2, num_hidden)
        self.relu_fn = nn.ReLU()
        self.linear2 = nn.Linear(num_hidden, num_outputs)

    def forward(self, text_data, image_data):
        text_output = self.text_classifier(text_data)
        image_output = self.image_classifier(image_data)
        x = torch.cat((text_output, image_output), dim=1)
        x = self.linear1(x)
        x = self.relu_fn(x)
        x = self.linear2(x)
        return x

```
