```python
file = r"Downloads\labells.csv.xls"
gh = pd.read_csv(file)
gh
```

```python
gh['text_corrected']
```

```python
from nltk.tokenize import RegexpTokenizer

punctuated = []
for i in range(len(gh)):
    tokenizer = RegexpTokenizer(r'\w+')
    b = tokenizer.tokenize(str(gh['text_corrected'][i]))
    a = ' '.join(b).lower()
    punctuated.append(a)
```

```python
punctuated
```

```python
text=pd.DataFrame()
text['image'] = gh['image_name']
text['Sentences'] = punctuated
text['Labels'] = gh['overall_sentiment']
d={'positive':1,'negative':-1, 'neutral':0 , 'very_positive': 1 , 'very_negative':-1}
text['labels']=text['Labels'].map(d)
```

```python
labels=np.array(text['labels'])
x=np.unique(labels)
```

```python
X = text.Sentences
y = text.Labels
```

```python
print(y)
```

```python
vectorizer = TfidfVectorizer()
X_trainn = vectorizer.fit_transform(X)
#X_testt = vectorizer.transform(y)
```

```python
#le = LabelEncoder()
#y = le.fit_transform(y)
X_train, X_test, y_train, y_test = train_test_split(X_trainn.toarray(), text['labels'], random_state=51, test_size=0.3)
```

```python
clf1 = DecisionTreeClassifier()
clf1.fit(X_train,y_train)
y_pred1=clf1.predict(X_test)
print('Accuracy:', accuracy_score(y_test, y_pred1))
print("F1 Score: ",f1_score(y_test, y_pred1,average='macro')) 
print(classification_report(y_test, y_pred1))
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf2=KNeighborsClassifier(algorithm='auto', leaf_size=30, metric='minkowski',metric_params=None, n_jobs=60, n_neighbors=5, p=2,weights='uniform')
bal=['ball_tree', 'kd_tree', 'brute']
clf2.fit(X_train, y_train)
y_pred2=clf2.predict(X_test)
print("Accuracy:", accuracy_score (y_test, y_pred2))
print("F1-Score Macro:", f1_score(y_test, y_pred2,average='macro'))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
clf3 = LogisticRegression(C=200, penalty='l1', solver='saga',n_jobs = 60)
clf3.fit(X_train,y_train)
y_pred3=clf3.predict(X_test)
print('Accuracy:', accuracy_score(y_test, y_pred3))
print("F1 Score: ",f1_score(y_test, y_pred3,average='macro')) 
print(classification_report(y_test, y_pred3))
#logModel_grid = GridSearchCV(estimator=LogisticRegression(random_state=1234), verbose=1, cv=10, n_jobs=-1)
#logModel_grid.fit(X_train, y_train)
#print(logModel_grid.best_estimator_)
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
print('Accuracy:', accuracy_score(y_test, y_pred))
print("F1 Score: ",f1_score(y_test, y_pred,average='macro')) 
print(classification_report(y_test, y_pred))
```

```python
Names =[]
Lens = []
ImData = list()
c = 0
for base in os.listdir(folder):
    print(c)
    Names.append(base)
    var = imread(folder + base, as_gray = True)
    var = skimage.transform.resize(var, (100,100))
    ImData.append(var)
    c+=1

```

```python
d={'positive':1,'negative':-1, 'neutral':0 , 'very_positive':1 , 'very_negative':-1}
df['label']=df['label'].map(d)
y = df['label']
```

```python
x_train, x_test, y_train, y_test = train_test_split(list(df['pixel']),df['label'] ,test_size = 0.3, random_state=51)
```

```python
clf4 = RandomForestClassifier()
clf4.fit(x_train,y_train)
y_pred4=clf4.predict(x_test)
print('Accuracy:', accuracy_score(y_test, y_pred4))
print("F1 Score: ",f1_score(y_test, y_pred4,average='macro')) 
print(classification_report(y_test, y_pred4))
```

```python
clf5 = MLPClassifier()
clf5.fit(x_train,y_train)
y_pred5=clf5.predict(x_test)
print('Accuracy:', accuracy_score(y_test, y_pred5))
print("F1 Score: ",f1_score(y_test, y_pred5,average='macro')) 
print(classification_report(y_test, y_pred5))
```

```python
clf6 = svm.SVC(C = 3)
clf6.fit(x_train,y_train)
y_pred6=clf6.predict(x_test)
print('Accuracy:', accuracy_score(y_test, y_pred6))
print("F1 Score: ",f1_score(y_test, y_pred6,average='macro')) 
print(classification_report(y_test, y_pred6))
```

```python
clf7 = BaggingClassifier()
clf7.fit(x_train,y_train)
y_pred7=clf7.predict(x_test)
print('Accuracy:', accuracy_score(y_test, y_pred7))
print("F1 Score: ",f1_score(y_test, y_pred7,average='macro')) 
print(classification_report(y_test, y_pred7))
```

```python
clf8 = HistGradientBoostingClassifier(loss='auto', learning_rate=0.1, max_iter=100, max_leaf_nodes=31,)
clf8.fit(x_train,y_train)
y_pred8=clf8.predict(x_test)
print('Accuracy:', accuracy_score(y_test, y_pred7))
print("F1 Score: ",f1_score(y_test, y_pred7,average='macro')) 
print(classification_report(y_test, y_pred7))
```
